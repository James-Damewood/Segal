```c++
int Simulate(int num,double curr[101]) {
    double field = 0.0;

    double N = 10.0;
    double N2 = 100.0;
    int N_int = 10;
    double f = 1.0;

    map<int, double> Energies_Stored;

    map<int, int> IndE;

    double Latt[10][10];

    for (int i = 0; i<10; i++) {
        for (int j = 0; j < 10; j++) {
            if (i == 4) {
                Latt[i][j] = 1.0;
            } else {
                Latt[i][j] = 1.0;
            } 
        }
    }

    double Ene = 0.0;
    double mag = 0.0;
    double WF = 0.0;
    for (int i = 0; i<10; i++) {
        for (int j = 0; j < 10; j++) {

            WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
            Ene = Ene - WF*Latt[i][j];
        }
    }

    cout << Ene << "\n";
    Ene = Ene/2.0;
    int tempE = round(Ene*10.0);
    cout << tempE <<"\n";

    Energies_Stored[tempE] = 0.0;

    int N_init = 1000000;
    int N2_int = 100;

    for (int k = 0; k < N_init; k++) {
        int i = int(rand()%N_int);
        int j = int(rand()%N_int);


        double S = Latt[i][j];    

        double WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
        double dmag = -2*Latt[i][j];
        double chpe = 0.0;

        double Enew = Ene + 2*S*WF +chpe;



        if (Energies_Stored.find(round(Enew*10.0))==Energies_Stored.end()){
            Energies_Stored[round(Enew*10.0)] = 0.0;
        } 

        double P = exp(Energies_Stored[round(Ene*10.0)]-Energies_Stored[round(Enew*10.0)]);  
        if ( P > (double) rand() / (RAND_MAX) ){
            Latt[i][j] = -1.0* Latt[i][j];
            Ene = Enew;   

        }

        Energies_Stored[round(Ene*10.0)] += f;


    }

    int inc = 0;


    for ( const auto &myPair : Energies_Stored ) {
        double curr = myPair.first;
        IndE[curr] = inc;
        inc = inc + 1;

    }

    cout << IndE[0] << "\n";;
        
    double lngE[99];  
    for (int i = 0; i < 99; i++) 
    {
        lngE[i] = 0.0;
    }

    int Hist[99];  
    for (int i = 0; i < 99; i++) 
    {
        Hist[i] = 0;
    }

    double Energies[99];

    inc = 0;
    for ( const auto &myPair : Energies_Stored ) {

        Energies[inc] = myPair.first/10.0;

        inc = inc + 1;

    }

    cout << Energies[49] << "\n";

    for (int i = 0; i<10; i++) {
        for (int j = 0; j < 10; j++) {
            if (i % 2 == 0) {
                Latt[i][j] = 1.0;
            } else {
                Latt[i][j] = -1.0;
            }
        }
    }

    Ene = 0.0;
    mag = 0.0;
    WF = 0.0;
    for (int i = 0; i<10; i++) {
        for (int j = 0; j < 10; j++) {

            WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
            //cout << WF << "\n";
            Ene = Ene - WF*Latt[i][j];
        }
    }

    cout << Ene << "\n";
    Ene = Ene/2.0;
    tempE = round(Ene*10.0);
    cout << tempE <<"\n";
    cout << mag << "\n";

    bool mode = false;
    double itt = 1;
    f = 1.0;
    while (itt < pow(10,num)){

        int i = int(rand()%N_int);
        int j = int(rand()%N_int);  // The coordinates of the site
        int S = Latt[i][j];          

        double WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
        double dmag = -2*Latt[i][j];
        double chpe = 0.0;

        double Enew = Ene + 2*S*WF +chpe;      // The energy of the tryed step


        double P = exp(lngE[IndE[round(Ene*10.0)]]-lngE[IndE[round(Enew*10.0)]]);  
        if ( P > (double) rand() / (RAND_MAX) ){
            Latt[i][j] = -1.0* Latt[i][j];
            Ene = Enew;   

        }

        Hist[IndE[round(Ene*10.0)]] += 1;  // Histogram is update at each Monte Carlo step!
        lngE[IndE[round(Ene*10.0)]] += f; // Density of states is also modified at each step!


        if (int(itt) % 99*1000 == 0 && mode == false){

            if (*min_element(Hist,Hist+99) > 0 && mode == false){



                for (int i = 0; i < 99; i++) // ...initialize it
                {
                    Hist[i] = 0;
                }

                f = f/2.0;

                if (  f <= (1.0 / (itt/99.0))    ){
                    mode = true;
                    f = (1.0 / (itt/99.0));
                }
            }

        }

        if (mode == true) {

            f = 1.0 / (itt/99.0);
        }





        //if (int(itt) % 1000000 == 0) {
        //    cout << f << "\n";
        //}

        itt += 1;
    }
    
    cout << f << "\n";

    cout << lngE[98] << "\n";
    
    double diff = lngE[0] - log(2.0);

    for (int i = 0; i <99; i++) {
        lngE[i] -= diff;
    }

    double temps_out[101];

    for (int i = 0; i < 101;i++){
        temps_out[i] = 1.5 + (2.0)*(float(i)/100.0);
    }


    cout.precision(10);
    
    double Z = 0.0;
    for (int i = 0; i < 101;i++){
        Z = 0.0;
        
        for (int j = 0; j < 99; j++ ) {
            Z += exp(lngE[j] - Energies[j]/temps_out[i]);
        }

        //cout << ",";
        curr[i] = -1.0*temps_out[i]*log(Z)/100.0;
    }
    
    
    
    return 0;
}
```
