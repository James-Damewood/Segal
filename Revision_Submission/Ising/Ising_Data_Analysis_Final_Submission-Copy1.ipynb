```python
def get_error(arr):
    
    errors = []
    
    for i in range(arr.shape[0]):
        curr_error = 0.0
        curr_error += np.abs(arr[i,0] - Exact_Ising[0])
        curr_error += np.abs(arr[i,25] - Exact_Ising[1])
        curr_error += np.abs(arr[i,50] - Exact_Ising[2])
        curr_error += np.abs(arr[i,75] - Exact_Ising[3])
        curr_error += np.abs(arr[i,100] - Exact_Ising[4])
        
        curr_error /= 5.0
        errors.append(curr_error)
    
    
    errors = np.asarray(errors)
    return errors
```

```python
hfont = {'fontname':'Arial'}
plt.rcParams["figure.figsize"] = (7,5)

c = newcmp(0.5)


plt.boxplot(np.log10(get_error(Ref_00[:,0,:])),positions=[np.log10(1000000)],patch_artist=True,
            boxprops=dict(facecolor="white", color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )

plt.boxplot(np.log10(get_error(Ref_00[:,1,:])),positions=[np.log10(10000000)],patch_artist=True,
            boxprops=dict(facecolor="white", color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )

plt.boxplot(np.log10(get_error(Ref_00[:,2,:])),positions=[np.log10(100000000)],patch_artist=True,
            boxprops=dict(facecolor="white", color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )

plt.boxplot(np.log10(get_error(Ref_00[:,3,:])),positions=[np.log10(1000000000)],patch_artist=True,
            boxprops=dict(facecolor="white", color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )

plt.boxplot(np.log10(get_error(Ref_00[:,4,:])),positions=[np.log10(10000000000)],patch_artist=True,
            boxprops=dict(facecolor="white", color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )

plt.boxplot(np.log10(get_error(Ref_00[:,5,:])),positions=[np.log10(100000000000)],patch_artist=True,
            boxprops=dict(facecolor="white", color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )


#plt.scatter(np.ones(50)*np.log10(10000000000),np.log10(get_error(Free_Energies_ML[0,:,:].T)),color=newcmp(0.5),alpha=0.3,label="Autoregressive")

c = 'black'

plt.boxplot(np.log10(get_error(Free_Energies_ML[0,:,:].T)),positions=[np.log10(46000000)],patch_artist=True,
            boxprops=dict(facecolor="white", color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),
            )

x_fake = np.linspace(1,2,10)
plt.plot(x_fake,x_fake,color='black',label="Autoregressive")
plt.plot(x_fake,x_fake,color=newcmp(0.5),label="Wang-Landau")

plt.xlim([5.5,11.5])
plt.ylim([-6.0,-1.5])

plt.xlabel('Log Energy Evaluations',fontsize=16,fontname="Arial")
plt.ylabel('Log Free Energy Error',fontsize=16,fontname="Arial")
plt.xticks([6,7,8,9,10,11])
plt.xticks(fontsize = 16,fontname = "Arial") 
plt.yticks(fontsize = 16,fontname = "Arial") 

font = font_manager.FontProperties(family='Arial',size=12)

plt.legend(prop=font)

plt.savefig("Timing_Final_Submission.pdf",bbox_inches='tight')

plt.show()
    
```
