```c++
map<int, double> Energies_Stored;

map<int, int> IndE;

double Latt[10][10];
```

```c++

for (int i = 0; i<10; i++) {
    for (int j = 0; j < 10; j++) {
        if (i == 4) {
            Latt[i][j] = 1.0;
        } else {
            Latt[i][j] = 1.0;
        }
        
        
    }
}

for (int i = 0; i<10; i++) {
    for (int j = 0; j < 10; j++) {
        cout << Latt[0][5] << "\n";
    }
}
```

```c++
double Ene = 0.0;
double mag = 0.0;
double WF = 0.0;
for (int i = 0; i<10; i++) {
    for (int j = 0; j < 10; j++) {
        mag += Latt[i][j];
        WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
        cout << WF << "\n";
        Ene = Ene - WF*Latt[i][j];
    }
}

cout << Ene << "\n";
Ene = Ene/2.0 - field*mag;
int tempE = round(Ene*10.0);
cout << tempE <<"\n";
cout << mag;
```

```c++
Energies_Stored[tempE] = 0.0;
```

```c++
int N_init = 1000000;
int N2_int = 100;

for (int k = 0; k < N_init; k++) {
    int i = int(rand()%N_int);
    int j = int(rand()%N_int);
    

    double S = Latt[i][j];    
   
    double WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
    double dmag = -2*Latt[i][j];
    double chpe = -field*dmag;
        
    double Enew = Ene + 2*S*WF +chpe;
        
   
    
    
        
    if (Energies_Stored.find(round(Enew*10.0))==Energies_Stored.end()){
        Energies_Stored[round(Enew*10.0)] = 0.0;
    } 
            
    double P = exp(Energies_Stored[round(Ene*10.0)]-Energies_Stored[round(Enew*10.0)]);  
    if ( P > (double) rand() / (RAND_MAX) ){
        Latt[i][j] = -1.0* Latt[i][j];
        Ene = Enew;   
        
    }
            
    Energies_Stored[round(Ene*10.0)] += f;
    
            
}
```

```c++
double lngE[514];  // declare the array and then...
for (int i = 0; i < 514; i++) // ...initialize it
{
    lngE[i] = 0.0;
}

int Hist[944];  // declare the array and then...
for (int i = 0; i < 514; i++) // ...initialize it
{
    Hist[i] = 0;
}
```

```c++
double Ene = 0.0;
double mag = 0.0;
double WF = 0.0;
for (int i = 0; i<10; i++) {
    for (int j = 0; j < 10; j++) {
        mag += Latt[i][j];
        WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
        cout << WF << "\n";
        Ene = Ene - WF*Latt[i][j];
    }
}

cout << Ene << "\n";
Ene = Ene/2.0 - field*mag;
int tempE = round(Ene*10.0);
cout << tempE <<"\n";
cout << mag;
```

```c++
bool mode = false;
double itt = 1;
double f = 1.0;
while (itt < 100000000000.0){
    
    int i = int(rand()%N_int);
    int j = int(rand()%N_int);  // The coordinates of the site
    int S = Latt[i][j];          
   
    double WF = Latt[(i+1)%N_int][j] + Latt[i][(j+1)%N_int] + Latt[(i-1+10)%N_int][j] + Latt[i][(j-1+10)%N_int];
    double dmag = -2*Latt[i][j];
    double chpe = -field*dmag;
        
    double Enew = Ene + 2*S*WF +chpe;      // The energy of the tryed step
    
    
    double P = exp(lngE[IndE[round(Ene*10.0)]]-lngE[IndE[round(Enew*10.0)]]);  
    if ( P > (double) rand() / (RAND_MAX) ){
        Latt[i][j] = -1.0* Latt[i][j];
        Ene = Enew;   
        
    }
        
    Hist[IndE[round(Ene*10.0)]] += 1;  // Histogram is update at each Monte Carlo step!
    lngE[IndE[round(Ene*10.0)]] += f; // Density of states is also modified at each step!
    
    
    if (int(itt) % 514*1000 == 0 && mode == false){
        
        if (*min_element(Hist,Hist+514) > 0 && mode == false){
            
            
            
            for (int i = 0; i < 514; i++) // ...initialize it
            {
                Hist[i] = 0;
            }
            
            f = f/2.0;
            
            if (  f <= (1.0 / (itt/514.0))    ){
                mode = true;
                f = (1.0 / (itt/514.0));
            }
        }
        
    }
        
    if (mode == true) {
            
        f = 1.0 / (itt/514.0);
    }
        
    
    
    
    
    //if (int(itt) % 1000000 == 0) {
    //    cout << f << "\n";
    //}
    
    itt += 1;
}
```
